```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./UNSW_NB15_training-set.csv')
```

```python
data
```

```python
data = data.drop('id', axis=1)
```

```python
data['label'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
data['proto'] = labelencoder.fit_transform(data['proto'])
data['state'] = labelencoder.fit_transform(data['state'])
data['service'] = labelencoder.fit_transform(data['service'])
data['attack_cat'] = labelencoder.fit_transform(data['attack_cat'])


data.head(5)
```

```python
data.info()
```

```python
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.model_selection import train_test_split


X = data.drop('attack_cat', axis=1) 
Y = data['attack_cat']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Select top k features using Chi-squared
k = 10  # Choose the desired number of features
chi2_selector = SelectKBest(chi2, k=k)
X_train_new = chi2_selector.fit_transform(X_train, Y_train)
X_test_new = chi2_selector.transform(X_test)

# Print the selected features
selected_features = [i for i in range(len(chi2_selector.get_support())) if chi2_selector.get_support()[i]]
print(f"Selected features: {selected_features}")
selected_columns = data.iloc[:, selected_features]
```
