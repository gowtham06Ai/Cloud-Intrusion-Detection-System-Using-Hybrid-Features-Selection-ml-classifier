```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('./dataset.csv')
```

```python
data
```

```python
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
data['protocol_type'] = labelencoder.fit_transform(data['protocol_type'])
data['flag'] = labelencoder.fit_transform(data['flag'])
data['service'] = labelencoder.fit_transform(data['service'])
data['Attack Type'] = labelencoder.fit_transform(data['Attack Type'])
data['target'] = labelencoder.fit_transform(data['target'])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.drop(labels=['Attack Type'], axis=1),
    data['Attack Type'],
    test_size=0.3,
    random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.feature_selection import mutual_info_classif
# determine the mutual information
mutual_info = mutual_info_classif(X_train, y_train)
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False)
```

```python
mutual_info.sort_values(ascending=False).plot.bar(figsize=(20, 8))
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
sel_five_cols = SelectKBest(mutual_info_classif, k=5)
sel_five_cols.fit(X_train, y_train)
X_train.columns[sel_five_cols.get_support()]
```
